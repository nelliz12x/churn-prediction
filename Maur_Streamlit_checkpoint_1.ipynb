```python
# Import all necessary libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Handle Missing and corrupted values

df.isnull().sum()
```

```python
# Deal with missing values accross all columns using mode and fillna
for i in df.columns: # Loop through the dataframe columns
  if df[i].isna().sum() > 0: # Check for missing values in each columns,
    df[i].fillna(df[i].mode()[0], inplace=True) # If missing value in column, use mode to fill the missing values

df.isnull().sum() # All missing value cleared
```

```python
# Using Label Encoding, change categorical data to numerical data
from sklearn.preprocessing import LabelEncoder # Import package label encoder
le = LabelEncoder()
col = [1, 2, 16] # COlumn index that are categorical data
for i in df.columns:
  if df[i].dtypes() == 'object':# Loop through the index
    df.iloc[:, i] = le.fit_transform(df.iloc[:, i]) # Fit and transform the column

df.head() # Show the data after label encoding
```

```python
X = df.drop(['MRG', 'user_id', 'CHURN'], axis=1) # Drop the columns that are not needed, and the target column as well
X.head()
```

```python
# There are high variation in our data, and we would need to standardize it to ensure we have good accuracy for our model
from sklearn.preprocessing import StandardScaler # Import package for standard scaler
sc = StandardScaler() # Create object for standard scaler
X = sc.fit_transform(X) # Fit and transform the data
X # Show the data after standard scaler
```

```python
y = df.CHURN.values # Get the target column
y # Show the target column
```

```python
from sklearn.model_selection import train_test_split # Import package for train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # Split the data into train and test
```

```python
from sklearn.linear_model import LogisticRegression # Import package for logistic regression
model = LogisticRegression(random_state=0) # Create object for logistic regression
model.fit(X_train, y_train) # Fit the data to the model
```

```python
pred = model.predict(X_test) # Predict the test data
pred # Show the prediction
```

```python
from sklearn.metrics import confusion_matrix # Import package for confusion matrix
cm = confusion_matrix(y_test, pred) # Create object for confusion matrix
cm # Show the confusion matrix
```

```python
from sklearn.metrics import classification_report # Import package for classification report
print(classification_report(y_test, pred)) # Show the classification report
```

```python
from sklearn.metrics import accuracy_score # Import package for accuracy score
print("Test Accuracy: ", accuracy_score(y_test, pred)) # Show the accuracy score for test data
print("Train Accuarcy: ", accuracy_score(y_train, model.predict(X_train))) # Show the accuracy score for train data
```
